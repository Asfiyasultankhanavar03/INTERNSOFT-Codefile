```python
variables = pd.DataFrame({'normal':np.random.normal(size=20),
                         'gamma':np.random.normal(size=20),
                         'poisson':np.random.normal(size=20)})
print(variables)
variables.cumsum(0).plot()
```

```python
variables.cumsum(0).plot(subplots=True)
```

```python
variables.cumsum(0).plot(subplots=False)
```

```python
variables.cumsum(0).plot(secondary_y='normal')
```

```python
variables.cumsum(0).plot(secondary_y='gamma')
```

```python
variables.cumsum(0).plot(secondary_y='poisson')
```

```python
fig, axes = plt.subplots(nrows=4, ncols=1, figsize=(4, 12))
for i,var in enumerate(['normal','gamma','poisson']):
    variables[var].cumsum(0).plot(ax=axes[i], title=var)
axes[0].set_ylabel('cumulative sum')
```
